```python
dasturchilar = {"Ali", "Vali", "G'ani", "Hasan", "Doniyor"}
kitobxonlar = {"Vali", "G'ani", "Murod", "Fotima", "Ali"}
kinochilar = {"Hasan", "Doniyor", "Murod", "Zuhra"}
```

```python
# Birlashma (Union)
# Savol: Umuman olganda, ushbu uchta klubdan kamida bittasiga a'zo bo'lgan barcha talabalar kimlar? 
# (Takrorlanmas ro'yxatni olish kerak).
barcha_azolar = dasturchilar.union(kitobxonlar).union(kinochilar)
barchasi = dasturchilar | kitobxonlar| kinochilar
barchasi
barcha_azolar
```

```python
# eng kami 2 ta guruhga a'zo bo'lganlar
dastur_kitob = dasturchilar & kitobxonlar
dastur_kino = dasturchilar & kinochilar
kino_kitob = kinochilar & kitobxonlar

ikkita_guruh = dastur_kitob | dastur_kino| kino_kitob
print(f"Kamida ikkitasiga a'zo bo'lganlar:{ikkita_guruh}")
```

```python
#3. Uchta Klubning Kesishmasi (Intersection)
# Savol: Uchta klubning hammasiga a'zo bo'lgan eng "faol" talabalar kimlar?
uchala = dasturchilar & kitobxonlar & kinochilar
uchala
```

```python
# Savol: Faqat Dasturlash Klubiga a'zo bo'lgan talabalar kimlar? 
#(Ya'ni, ular kitob yoki kino klubiga a'zo emas)
# dasturchilar - (kitobxonlar + kinochilar)
kitob_kino = kitobxonlar | kinochilar
faqat_dasturchilar = dasturchilar - kitob_kino
faqat_dasturchilar
```

```python
#Kitobxonlarni topamiz, ya'ni faqat kitobxonlarga a'zo bo'lgan
kino_dasturchi = dasturchilar | kinochilar
faqat_kitob = kitobxonlar - kino_dasturchi
print(f"{faqat_kitob} faqat kutubxonaga a'zo ")
```

```python
# faqat dasturchilik yoki kitobxonlik kulubiga a'zo bo'lganlar kimlar?
#(ikkisida bori chiqariladi)
faqat_bir = dasturchilar ^ kitobxonlar
faqat_bir
```

```python
# dictionary
# Vazifa: dict yordamida talaba ma'lumotlarini saqlang (ism, yosh, baholar).
# Funksiya yozing, u talabaning o'rtacha bahosini hisoblaydi yoki yangi ma'lumot qo'shadi. 
# Loop yordamida barcha talabalarni chop eting.
def talaba_malumotlari(talabalar, talaba_id, amal, yangi_malumot=None):
    if amal == "o'rtacha_baho":
        baholar = talabalar[talaba_id]["baholar"]
        ortacha = sum(baholar) / len(baholar)
        print(f"{talabalar[talaba_id]['ism']}ning o'rtacha bahosi: {ortacha}")
    elif amal == "qo'shish" and yangi_malumot:
        talabalar[talaba_id] = yangi_malumot
        print(f"{talaba_id} ID bilan talaba qo'shildi.")
    
    for id, info in talabalar.items():
        print(f"ID: {id}, Ma'lumot: {info}")

# Test
talabalar = {
    101: {"ism": "Ali", "yosh": 20, "baholar": [85, 90, 78]},
    102: {"ism": "Vali", "yosh": 21, "baholar": [88, 92, 80]}
}
talaba_malumotlari(talabalar, 101, "o'rtacha_baho")
talaba_malumotlari(talabalar, 103, "qo'shish", {"ism": "Nodira", "yosh": 19, "baholar": [90, 95, 88]})
```

```python
# talabalarni baholarini tahlil qiladigan dastur yozamiz
def tahlil(baho_lugati):
    """ talabalarni baholarini tahlil qiladi, eng yuqori va past bahoni qaytaradi"""
    if not baho_lugati:
        print("Iltimos ro'yxat kiriting")
    # barcha baholarni olish
    faqat_baholar = baho_lugati.values()

    # eng kattasi va kichkinasini topish
    eng_katta = max(faqat_baholar)
    eng_kichkina  = min(faqat_baholar)

    # shu bahoga ega odamlar uchun ro'yxat qilamiz
    eng_yaxshilar = []
    eng_yomonlar = []

    # endi loop orqali bahosi qaysinga tushsa shuni ko'rsatamiz va yigamiz
    for ism, baho in baho_lugati.items():
        if baho == eng_katta:
            eng_yaxshilar.append(ism)
        elif baho== eng_kichkina:
            eng_yomonlar.append(ism)
    natija ={
        "Eng yuqori baho": eng_katta,
        "Eng yuqori bahoga ega": eng_yaxshilar,
        "Eng past baho": eng_kichik,
        "Eng past bahoga ega": eng_yomonlar}
    return natija
    
```
