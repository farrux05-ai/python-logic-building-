```python
# ============================================
# 4. LIST ICHIDA DICT ICHIDA LIST (3 QAVAT!)
# ============================================
print("\n4. LIST ICHIDA DICT ICHIDA LIST - 3 QAVAT")
print("=" * 60)
maktab = [
    {
        "sinf": "10-A",
        "talabalar": ["Ali", "Vali"],
        "o'qituvchi": "Aziz Aka"
    },
    {
        "sinf": "10-B", 
        "talabalar": ["Sami", "Jamila", "Nodira"],
        "o'qituvchi": "Malika Opa"
    }
]
print("Struktura (soddalashtirilgan):")
print("maktab = [")
print("  {sinf: '10-A', talabalar: ['Ali', 'Vali']},")
print("  {sinf: '10-B', talabalar: ['Sami', 'Jamila', 'Nodira']}")
print("]")
print()

print("PYTHON QANDAY O'QIYDI - QADAM BA QADAM:")
print()

print("QAVAT 1: List ustida loop (har bir sinf)")
for sinf_index, sinf_data in enumerate(maktab):
    print(f"\n{'*' * 20}")
    print(f"Qadam {sinf_index + 1}: maktab[{sinf_index}]")
    print(f"  Bu nima? Dictionary: {type(sinf_data).__name__}")
    print(f"  Qiymat: {sinf_data}")
    
    print(f"\n  QAVAT 2: Dictionary ichida (sinf ma'lumotlari)")
    sinf_nomi = sinf_data['sinf']
    oqituvchi = sinf_data["o'qituvchi"]
    talabalar = sinf_data['talabalar']
    
    print(f"    Sinf: {sinf_nomi}")
    print(f"    O'qituvchi: {oqituvchi}")
    print(f"    Talabalar: {talabalar}")
    
    print(f"\n    QAVAT 3: Talabalar listi ustida loop")
    for t_index, talaba in enumerate(talabalar):
        print(f"      [{t_index}] {talaba}")
        print(f"          Yo'l: maktab[{sinf_index}]['talabalar'][{t_index}]")

print("\n" + "=" * 60)
```

```python
# ============================================
# 7. XAVFSIZ ISHLASH - XATOLIKLARDAN HIMOYA
# ============================================
print("\n7. XAVFSIZ ISHLASH")
print("=" * 60)

# Buzilgan ma'lumot
buzilgan_data = [
    {
        "ism": "Ali",
        "loyihalar": [
            {"nomi": "Web sayt"}  # 'holat' kaliti yo'q!
        ]
    },
    {
        "ism": "Vali"
        # 'loyihalar' kaliti yo'q!
    }
]
print("Xavfsiz usul - KeyError'dan himoya:")
for xodim in buzilgan_data:
    ism = xodim.get('ism', 'Noma\'lum')
    print(f"\nXodim: {ism}")
    
    # get() metodi - agar kalit bo'lmasa None qaytaradi
    loyihalar = xodim.get('loyihalar', [])
    
    if loyihalar:
        for loyiha in loyihalar:
            nomi = loyiha.get('nomi', 'Nomi yo\'q')
            holat = loyiha.get('holat', 'Ma\'lumot yo\'q')
            print(f"  - {nomi}: {holat}")
    else:
        print(f"  (Loyihalar yo'q)")

print("\n" + "=" * 60)
```

```python
# ============================================
# 8. Get(), Try-expect
# ===========================================
print("=" * 70)
print("1. GET METODI - DICTIONARY'DAN XAVFSIZ QIYMAT OLISH")
print("=" * 70)
talaba = {
    "ism": "Ali",
    "yosh": 25,
    "maosh": 5000000
}
print("\n ODDIY USUL (Xavfli):")
print("-" * 70)
print("Kod: talaba['ism']")
print(f"Natija: {talaba['ism']}")

print("\nKod: talaba['kasb']  # Bu kalit yo'q!")
try:
    print(talaba['kasb'])
except KeyError as e:
    print(f"Xatolik! {e} topilmadi")
    
print("\n GET METODI (Xavfsiz):")
print("-" * 70)
print("get() metodi 2 parametr qabul qiladi:")
print("  1. Kalit (majburiy)")
print("  2. Default qiymat (ixtiyoriy) - agar kalit bo'lmasa qaytaradi")

print("\n1Ô∏è.  Oddiy get - faqat kalit:")
kasb = talaba.get('kasb')
print(f"Kod: talaba.get('kasb')")
print(f"Natija: {kasb}")
print(f"Turi: {type(kasb)}")
print("‚úì Xatolik yo'q, None qaytardi!")

print("\n2Ô∏è.  Get default qiymat bilan:")
kasb = talaba.get('kasb', 'Ma\'lumot yo\'q')
print(f"Kod: talaba.get('kasb', 'Ma\'lumot yo\\'q')")
print(f"Natija: {kasb}")
print("‚úì Default qiymat qaytardi!")

print("\n3Ô∏è.  Get mavjud kalit uchun:")
ism = talaba.get('ism', 'Noma\'lum')
print(f"Kod: talaba.get('ism', 'Noma\\'lum')")
print(f"Natija: {ism}")
print("‚úì Kalit mavjud bo'lgani uchun default qiymatni e'tibor bermadi!")

print("\n" + "=" * 70)

# GET METODI - AMALIY MISOLLAR
# ============================================
print("\n AMALIY MISOLLAR:")
print("-" * 70)

# Buzilgan ma'lumotlar
talabalar = [
    {"ism": "Ali", "yosh": 25, "ball": 85},
    {"ism": "Vali", "yosh": 30},  # ball yo'q
    {"ism": "Sami", "ball": 92},  # yosh yo'q
    {"yosh": 28, "ball": 78}      # ism yo'q
]
# Vazifa get metodidan foydalanib ism, yosh, ballni chiqarish kerak
for i, talaba in enumerate(talabalar, start=1):
    print(f"Talaba #{i}:")
    
    ism = talaba.get('ism', "Ma'lumot yo'q")
    yosh = talaba.get('yosh', "Yoshi berilmagan")
    ball = talaba.get('ball', "Ball topilmadi")

    print(f"Talabaning ismi: {ism}")
    print(f"Talabaning yoshi: {yosh}")
    print(f"Talabaning balli: {ball}")
    print()
print('=' * 70)

# ============================================
# 2. TRY-EXCEPT - ERROR HANDLING
# ============================================
print("\n2. TRY-EXCEPT - XATOLIKLARNI USHLASH")
print("=" * 70)

print("\nüìå TRY-EXCEPT SINTAKSISI:")
print("-" * 70)
print("""
try:
    # Xavfli kod (xatolik bo'lishi mumkin)
    natija = 10 / 0
except XatolikTuri as e:
    # Xatolik bo'lsa bu ishga tushadi
    print("Xatolik yuz berdi:", e)
""")

try:
    print("Kod: 10 / 2")
    natija = 10 / 2
    print(f"‚úì Natija: {natija}")
except ZeroDivisionError:
    print(" Nolga bo'lish mumkin emas!")

print()

try:
    print("Kod: 10 / 0")
    natija = 10 / 0
    print(f"Natija: {natija}")
except ZeroDivisionError:
    print(" Nolga bo'lish mumkin emas!")

print("\n" + "=" * 70)

print("\n2Ô∏è.  KO'P XATOLIKLARNI USHLASH:")
print("-" * 70)

data = [
    {"ism": "Ali", "yosh": "25"},     # yosh string (int bo'lishi kerak)
    {"ism": "Vali", "yosh": 30},      # to'g'ri
    {"ism": "Sami"},                  # yosh yo'q
    {"yosh": 28}                      # ism yo'q
]

for i, m in enumerate(data, start=1):
    print(f"Talaba {i} {m}haqida ma'lumotlar:")
    try:
        # bu yerda agar ism bo'lmasa KeyError beradi
        ism = m['ism']

        # bu yerda agar yosh intga aylantirsak ValueError berishi mumkinn
        yosh =int(m['yosh'])

        # bu yerda TypeError berishi mumkin
        yosh_10_yildan_keyin = yosh + 10

        print(f"  ‚úì {ism}: {yosh} yosh ‚Üí 10 yildan keyin {yosh_10_yildan_keyin}")

    except KeyError as e:
        print(f'KeyError: {e} topilmadi')
    except ValueError as e:
        print(f"  ValueError: Yoshni raqamga aylantirib bo'lmadi")
    except TypeError as e:
        print(f"  TypeError: {e}")

print("\n" + "=" * 70)

# Fayl o'quvchi dastur
def fayl_oqish(file_name):
    print(f"Fayl o'qilmoqda {file_name}")

    try:
        with open(file_name, 'r') as f:
            mazmun = f.read()
        print(mazmun)
        
    except FileNotFoundError:
        print(f"File topilmadi")
        mazmun = None
        
    except PermissionError:
        print("  Faylni ochishga ruxsat yo'q")
        
    else:
        print("   Hech qanday xatolik bo'lmadi")
    
    finally:
        print("   Bu har doim ishga tushadi (tozalash uchun)")

    return mazmun
# test
fayl_oqish("mavjud_emas.txt")
```
