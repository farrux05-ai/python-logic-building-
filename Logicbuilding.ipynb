```python
# ============================================
# LEVEL 6: GURUHLASH (GROUPING)
# ============================================
print("\n\n LEVEL 6: GURUHLASH")
print("-" * 70)

# MUAMMO 7: Savdolarni mijoz bo'yicha guruhlash
print("\n Muammo 7: Har bir mijoz qancha xarid qildi?")
print("-" * 70)

savdolar = [
    {"mijoz": "Ali", "mahsulot": "Telefon", "summa": 1500000},
    {"mijoz": "Dilnoza", "mahsulot": "Noutbuk", "summa": 3500000},
    {"mijoz": "Ali", "mahsulot": "Telefon", "summa": 150000},
    {"mijoz": "Dilnoza", "mahsulot": "Mouse", "summa": 50000},
    {"mijoz": "Ali", "mahsulot": "Noutbuk", "summa": 250000}
]

# Pattern: Guruhlash uchun dict ishlatamiz. Chunki bizga mijoz va summasi kerak('{Ali : 1000000}')
mijozlar = {}
for mijoz in savdolar:
    mijoz_ism = mijoz['mijoz']
    
    if mijoz_ism not in mijozlar:
        mijozlar[mijoz_ism] = mijoz['summa']
    else:
        mijozlar[mijoz_ism] +=(mijoz['summa'])
        
        
print(mijozlar)
mijoz_xaridlari = {}
for savdo in savdolar:
    mijoz = savdo['mijoz']
    summa = savdo['summa']

    if mijoz not in mijoz_xaridlari: # {'Ali': 0} # biz 0 qiymat qilib olamiz
        mijoz_xaridlari[mijoz] = 0

    mijoz_xaridlari[mijoz] += summa

print("\nMijozlarimiz haridlari va summalari: ")
for mijoz, summa in mijoz_xaridlari.items():
    print(f"Mijoz {mijoz} summa: {summa}")
    
# Deylik mana bunday ro'yxat qilsakchi{ Ali: [telefon, noutbuk]}
kategoriya_mijoz = {}

for savdo in savdolar:
    mijoz = savdo['mijoz']
    mahsulot = savdo['mahsulot']
    
    if mijoz not in kategoriya_mijoz: # agar mijoz ismmi kategoriya ichida bo'lmasa, yangi kalit va qiymat qo'sjadi aks holda qo'shib ketaveradi
        kategoriya_mijoz[mijoz] = [mahsulot]
        
    elif mahsulot not in kategoriya_mijoz[mijoz]: # bu yerda biz agar mahsulot kategoriya ichida bo'lmasa qo'shadi aks holda qoshmaydi
        kategoriya_mijoz[mijoz].append(mahsulot)
        
print(kategoriya_mijoz)



# shuni endi .get() metodi bilan qilamiz
mijozlar = {}
for mijoz in savdolar:
    mijoz_ism = mijoz['mijoz']
    mijozlar[mijoz_ism] = mijozlar.get(mijoz_ism, 0) + mijoz['summa']
print(mijozlar)

# .get() Bu yerda biz kategoriya orqali group qilamiz, mahsulot qancha sotildi
kategoriya_summa = {}
for kat in savdolar:
    kategoriya = kat['mahsulot']

    kategoriya_summa[kategoriya] = kategoriya_summa.get(kategoriya, 0) + kat['summa']
print("Kategoriya bo'yicha summa: ")
for kat, summ in kategoriya_summa.items():
    print(f"{kat} : {summ:,} so'm")

# default dict bilan qanday olish mumkin?
from collections import defaultdict

mijozlar = defaultdict(int)

for mijoz in savdolar:
    mijoz_ism = mijoz['mijoz']
    mijozlar[mijoz_ism] += mijoz['summa']
print(mijozlar)

# Bu yerda bizga set() duplikatlarsiz yig'ishga yordam berdi
mijozlar = defaultdict(set)

for mijoz in savdolar:
    mijoz_ism = mijoz['mijoz']
    mahsulot = mijoz['mahsulot']
    mijozlar[mijoz_ism].add(mahsulot)
print(mijozlar['Ali'])
```

```python
# ============================================
# LEVEL 9: MAX/MIN/SORTED
# ============================================
print("\n\n LEVEL 9: MAX/MIN/SORTED")
print("-" * 70)

# MUAMMO 10: Eng qimmat va eng arzon mahsulot
print("\nðŸŽ¯ Muammo 10: Eng qimmat va eng arzon mahsulotlar")
print("-" * 70)

# Qisqa va oddiy list comprehention bilan
narxlar = [m['narx'] for m in mahsulotlar]
print(f"Eng qimmat mahsulot narxi: {max(narxlar)}")
print(f"Eng arzon mahsulotlar narxi: {min(narxlar)}")

# Key bilan : to'liq ro'yxat 
eng_qimmat = max(mahsulotlar, key = lambda m: m['narx']) # bu yerda biz mahsulot haqida hamma ma'lumotlarni olamiz
eng_arzon = min(mahsulotlar, key = lambda m: m['narx'])

print(f"\n Eng qimmat mahsulot {eng_qimmat['nomi']} - {eng_qimmat['narx']:,} so'm - miqdor {eng_qimmat['miqdor']} ta")
print(f"\n Eng arzon mahsulot {eng_arzon['nomi']} - {eng_arzon['narx']:,} - miqdor {eng_arzon['miqdor']} ta")

# Narx bo'yicha qanday qilib tartiblab chiqarsak bo'ladi
tartiblangan = sorted(mahsulotlar, key = lambda m: m['narx'], reverse=True) # bizga to'liq ro'yxatni beradi
for tartib in tartiblangan:
    print(f"Mahsulotlar: {tartib['nomi']} - narxi {tartib['narx']}")
    
# miqdor bo'yicha tartiblash
tartib = sorted(mahsulotlar, key = lambda m: m['miqdor'], reverse=True)

for i, t in enumerate(tartib, 1):
    print(f"Mahsulotlar miqdor bo'yicha eng ko'p sotilgani {i} o'rin: {t['nomi']} - miqdor {t['miqdor']} ta")

# xatto lambda ichida hisoblab buni qaytarishimiz ham mumkin
tar = sorted(mahsulotlar, key = lambda m: m['miqdor'] * m['narx'], reverse=True)
tar

```

```python
# ============================================
# LEVEL 10: FILTER VA MAP
# ============================================
print("\n\nðŸ“š LEVEL 10: FILTER VA MAP")
print("-" * 70)

# MUAMMO 11: Faqat qimmat mahsulotlar (1,000,000+ so'm)
print("\nðŸŽ¯ Muammo 11: 10,000 so'mdan qimmat mahsulotlar")
print("-" * 70)

print('\n Sintaksis')
print("\nmap(function, data") # funskiyani hammasiga qo'llaydi
print("\nfilter(function, data)") # filter esa true qaytarganini saqlab qoladi
# oddiy loop bilan
qimmat = []
# Usul 1: Loop bilan
for m in mahsulotlar:
    if m['narx'] > 10000:
        qimmat.append(m)
print(f"10000 ming so'mdan mahsulot {qimmat}")\

# Usul 2: Filter bilan (professional)
filter_mah = list(filter(lambda m: m['narx'] > 10000, mahsulotlar))
for m in qimmat:
    print(f"\n  {m['nomi']}: {m['narx']:,} so'm")

# List comprehention bilan
qimmat1 = [m for m in mahsulotlar if m['narx'] > 10000]
print(qimmat1)

# MAP: Faqat nomlarni olish
print("\nFaqat nomlar (map bilan):")
nomlar = list(map(lambda m: m['nomi'], mahsulotlar))
print(f"  {', '.join(nomlar)}")

sonlar = [1, 2, 3, 4, 5, 6]
natija = list(
    map(lambda x: x ** 2,      # har birini kvadrat qilish
        filter(lambda x: x % 2 == 0, sonlar))  # faqat juftlarni tanlash
)

print(natija)  # [4, 16, 36]


    
```
