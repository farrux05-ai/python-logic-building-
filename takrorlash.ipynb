```python
import random
# o'yin tuzamiz
# tasodiy sonlar generatori
# 1. O'yinni sozlash
pastki_son = 1
yuqori_son = 100
tasodifiy_son  = random.randint(pastki_son, yuqori_son)
urinishlar_soni = 0
topildimi = False

print(f"======================================")
print(f"sonni topish o'yini")
print(f"{pastki_son} va {yuqori_son} o'rtasidan bir son o'yladim. shuni siz topasiz")

while not topildimi:# foydalanuvchi topmaguncha davom etadi. not false == True ya'ni tru  bo'lguncha
    try:
        taxmin = int(input("Taxminiy sonnni kiriting: "))
        urinishlar_soni += 1
    except ValueError:
        print("Iltimos son kiriting")
        continue # keyingiga o'tkazamiz
        
    # shartlarni tekshirish
    if taxmin > yuqori_son or taxmin < pastki_son:
        print(f"iltimos {pastki_son} va {yuqori_son} o'rtasidagi sonni kiriting")
        continue
    
    if taxmin > tasodifiy_son:
        print("Men o'ylagan son bundan kichik!")
        
    elif taxmin < tasodifiy_son:
        print("Men o'ylagan son bundan katta!")
    else:
        topildimi = True # bu yerga break desak ham bo'lardi
#4. Natijani chiqarish
print(f"Tabriklaymiz siz {urinishlar_soni} ta urinishda sonni topdingiz")
        
        
        
    


```
